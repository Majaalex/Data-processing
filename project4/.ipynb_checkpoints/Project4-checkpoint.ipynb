# 1. C-cord

The first task is to create 3 sine-waves using the frequencies 

###  Setting our frequencies and time scale

Our x is the amount of timestamps we have in our timeframe of 0 to. We make use of linspace to create these values.

````python
x = np.linspace(0,1,num=16000)
````

Our 3 frequencies, these are taken from https://www.szynalski.com/tone-generator/ and can be swapped out if you want to.

The 3 frequencies in order are c1, e3, g5

````python
c3 = 32.70319566257483
e3 = 164.81377845643496
g3 = 783.9908719634985
````

## Create our sine-waves

We want to create a sine-wave that is a combination of the 3 above frequencies. This means that we need to create 3 sine-waves which we then add together. Since we want our y-axis to stay between 1 and -1 we dampen the sine-waves by half so that the combination doesn't have clipping.

The formula we use is 
```
A * sin(2 * pi * time * frequency)
```
where A is the amplitude.

This formula is taken from https://physics.info/music/

We define a function so that we can just input a frequency to apply the above formula to the frequency to minimize typing.

Our amplitude is 1/3 since we will have 3 sine-waves and we want our combination of the 3 to be a sine-wave going from -1 to 1.

````python
def wave(freq):
    return 1 / 3 * np.sin(np.pi * 2 * x * freq)
````

Create our sine waves

````python
a = wave(c3)
b = wave(e3)
c = wave(g3) 
````

### Plot our 3 plots using plt

We make use of pyplot, and subplots to show our 3 sine-waves. We limit the graph to only show between 0 and 0.2 to have a better visual.

The rows and cols vaiables are use to define the size of our plot, so our plot will have space for 3 subplots, where each subplot is on a new row but in the same column.

````python
rows = 3
cols = 1
# X and y label names
ylab = 'Strength'
xlab = 'Timeframe'
# Titlee and label fontsize
t_size = 14
l_size = 12
# Plot dimensions
plt.figure(figsize=(14,8))

# c3 frequency wave
plt.subplot(rows,cols,1)
plt.title('Frequency: ' + str(c3), fontsize=t_size)
plt.xlabel(xlab, fontsize=l_size)
plt.ylabel(ylab, fontsize=l_size)
plt.plot(x, a)
plt.xlim(right=0.2, left=0)

# e3 frequency wave
plt.subplot(rows,cols,2)
plt.title('Frequency: ' + str(e3), fontsize=t_size)
plt.xlabel(xlab, fontsize=l_size)
plt.ylabel(ylab, fontsize=l_size)
plt.plot(x, b)
plt.xlim(right=0.2, left=0)

# g3 frequency wave
plt.subplot(rows,cols,3)
plt.title('Frequency: ' + str(g3), fontsize=t_size)
plt.xlabel(xlab, fontsize=l_size)
plt.ylabel(ylab, fontsize=l_size)
plt.plot(x, c)
plt.xlim(right=0.2, left=0)

plt.tight_layout()
plt.show()
````

## Combine the sine-waves to a C-chord

We can now simply add together the sine-waves to create our C-chord. We do this using the numpy add function, and then simply show it in the same way as above using pyplot.

Since we dampened our previus sine-waves, this one now has a y-scale going from -1 to y.

This graph is also shortened and only shows the values from 0 to 0.2 instead of 0 to 1.

````python
first = np.add(a,b)
combined = np.add(first, c)
plt.figure(figsize=(20,5))
plt.title('A C-chord made from the 3 previous sine-waves', fontsize=18)
plt.xlim(right=0.2, left=0)
plt.xlabel(xlab, fontsize=15)
plt.ylabel(ylab, fontsize=15)
plt.plot(x, combined)
plt.show()
````

We will output this since wave into the format sox wants from a dat file to be able to convert it to the .wav file. This format is 2 lines of comments with the sample rate and the channels.

We create the two dataframes to be able to make use of the pandas dataframe.to_csv function.

One contains our data values, x and y

The other our 2 comments for the dat file

# 2. Images

We take our image of a pokemon and we want to split it into the three luminance channels, r, g, b. To do this we make use of the matplotlib.image library.

We start by taking loading the image into the script. It takes the image from the same directory as the script.

````python
image = im.imread('venasaur.jpg')
````

To verify that it imported correctly we can show it off using plt.imshow

````python
plt.imshow(image)
plt.axis('off')
plt.show()
````

## Pick out the luminance channels

````python
img_1 = image[:, :, 0]
img_2 = image[:, :, 1]
img_3 = image[:, :, 2]
````

With these luminance channels, we can now draw these images using a grey colormap to show off how much of each color there is in each picture.

````python
plt.figure(figsize=(14,4))

# Red
plt.subplot(1,3,1)
plt.title('Red')
plt.axis('off')
plt.imshow(img_1, cmap='gray')

# Green
plt.subplot(1,3,2)
plt.title('Green')
plt.axis('off')
plt.imshow(img_2, cmap='gray')

# Blue
plt.subplot(1,3,3)
plt.title('Blue')
plt.axis('off')
plt.imshow(img_3, cmap='gray')


plt.show()
````

We now want to combined these 3 pictures into one. We can do this by creating a fresh np array with the correct dimensions. This was done by creating an array using np.indices with the images shape (which in this case is 64,64,3).
This placeholder is of the shape 3,64,64,3 so if we take the first object in the placeholder we end up with the format we wanted which is 64,64,3.

````python
placeholder = np.indices(image.shape)
print("full size: ", placeholder.shape)
print('First object: ', placeholder[0].shape)
recreation = placeholder[0]
````

With the correct format we can simply reverse what we did in the start and pick out each channel and set those channels to the luminance channels we pulled out of our original image.

````python
plt.imshow(recreation)
plt.axis('off')
plt.show()
````

````python
recreation[:,:,0] = img_1
plt.title('Added red channel')
plt.axis('off')
plt.imshow(recreation)
````

````python
recreation[:,:,1] = img_2
plt.title('Added green channel')
plt.axis('off')
plt.imshow(recreation)
````

````python
recreation[:,:,2] = img_3
plt.title('Added blue channel')
plt.axis('off')
plt.imshow(recreation)
````

With all the channels added to our recreation, we now have the same image as our original!

````python
plt.subplot(1,2,1)
plt.imshow(recreation)
plt.title('Recreation')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image)
plt.title('Original')
plt.axis('off')

plt.show()
````

# 3. Hidden alpha message
Here we want to add a hidden message by using the alpha channel in an image. We again start by loading in the image.

````python
fabio = im.imread('fabio64.png')
````

We currently have an image with the shape 64,64,3

This means we only have the r,g,b channels. We want to have an image with the r,g,b,a channels which means the shape would be 64,64,4.

We can create this in the same way as for our recreation.

````python
form = np.indices((64,64,4))
final = form[0]
final.shape
````

Now, since our image was a png this time, the values aren't the correct format so we fix that by multiplying by 255 and converting them to ints.

````python
fabio = fabio * 255
fabio = fabio.astype(int)
````

This lets us pick out the different channels of our fabio image

````python
fab1 = fabio[:,:,0]
fab2 = fabio[:,:,1]
fab3 = fabio[:,:,2]
````

We can add these channels to our freshlyl created empty image. This means we have a fabio image with an alpha channel

````python
final[:,:,0] = fab1
final[:,:,1] = fab2
final[:,:,2] = fab3
````

````python
plt.title('Fabio with an alpha channel')
plt.imshow(final)
plt.axis('off')
plt.show()
````

Now, this isn't entirely what we went out to do so we created a transparent image using gimp. The image is a text intended to have low transparency.

````python
alpha = im.imread('alpha.png')
alpha.shape
alpha = alpha * 255
alpha = alpha.astype(int)
````

To avoid overrites we create a copy of our image

````python
new = final
````

The picture we created ended up having "reversed" transparency values, so we fixed that by setting 0's to 255 and all other values respectively. Our images alpha channel is now updated with this new one.

````python
values = alpha[:,:,3]
values = 255 - values
new[:,:,3] = values
````

With that, we can now draw our images. We first draw our full image. It is seemingly normal. However if we take a look at the alpha channel, we can see that we do indeed have a hidden text.

````python
plt.figure(figsize=(14,4))

plt.subplot(1,2,1)
plt.title('Image with a hidden text')
plt.axis('off')
plt.imshow(new)

plt.subplot(1,2,2)
plt.title('The hidden text')
plt.axis('off')
plt.imshow(new[:,:,3])

plt.show()
````

We can visualize the fact that we have potentially hidden information in the alpha channel by creating a histogram of the alpha values. By doing this we can see that the image for the most part is 255 values (transparent pixels) however in this case there are plenty of values that are close to fully transparent but not quite.

````python
plt.hist(new[:,:,3].ravel(), bins=6)
plt.show()
````
